Dan buatlah daftar support dan confidence untuk 2 item set dengan menggunakan dataset tersebut (hint : buatlah tabel tabulasi terlebih dahulu untuk memudahkan)

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("dataset_market_basket.csv")
```

```python
df
```

```python
# 1 item set

df.sum()
```

```python

# 2 Item set
from itertools import permutations

#membuat perbandingan data
rules = list(permutations(df.columns, 2))
rules
```

```python
# mengfilter data yang dimana hanya item yang ada saja yang akan terhitung

rule_baru = []
for rule in rules:
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0:
    rule_baru.append(rule)
rule_baru
```

```python
# melihat jumlah data 2 item set
hitung = {} #Ganti dictionary
for rule in rule_baru:
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1)
      ])
  hitung[str(rule)] = banyak_data

hitung
```

```python
# List support
supp = [i/len(df)*100 for i in hitung.values()]
print(supp)
print("---")
for item,support in hitung.items():
  print(f"{item} support = {round(support/len(df)*100)}%")
```

```python
# List Confidence

print("---")
for item,ab in hitung.items():
  i = eval(item)
  print(f"{i[0]}->{i[1]} Confidence = {ab/df[i[0]].sum()*100}%")
```

1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi

```python
df2 = pd.read_csv("deteksi_uang_palsu.csv")
#dataframe sebelum ditambahkan kolom jarak
df2
```

```python
# 1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi
from sklearn.metrics.pairwise import euclidean_distances

#sebelum menambah jarak, pertama kita membuat dataframe baru berdasarkan data yang telah diberikan
data_baru = pd.DataFrame({'diagonal'    : [172.25],
                         'height_left' : [103.94],
                         'height_right': [103.68],
                         'margin_low'  : [4.14],
                         'margin_up'   : [3.25],
                         'length'      : [112.48]})

#kemudian kita menghitung jarak antara setiap baris dari df2 dengan data_baru
df2['distance'] = euclidean_distances(df2[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']], data_baru)
#dataframe setelah ditambahkan kolom jarak
df2
```

2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)

```python
# 2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)

k = int(np.sqrt(len(df2))) # mencari nilai K
data_terdekat = df2.sort_values('distance').head(k) # mengurutkan data

print(data_terdekat)
```

3. Tentukan apa hasil prediksinya berdasarkan label terbanyak

```python
# 3. Tentukan apa hasil prediksinya berdasarkan label terbanyak
from sklearn.neighbors import KNeighborsClassifier

# Menentukan data yang akan menjadi target
X_money = df2.drop(columns=['is_genuine','distance'],axis=1) #data
y_money = df2['is_genuine'] #target

# membuat data_baru sebagai sample
sample = data_baru

# membuat variabel knn yang dimana nilai k sebagai tuning parameter
knn = KNeighborsClassifier(n_neighbors=k)

# kemudian melatih data
knn.fit(X_money, y_money)


# memprediksi hasil
print("Maka hasilnya adalah : "+ ' '.join(map(str, knn.predict(sample))))
```

3. Data Preparation


```python
from sklearn.preprocessing import StandardScaler
df3 = df3.dropna()

# Handling outlier
cols = df3.columns[:-1]
for col in cols:
    q1 = df3[col].quantile(0.25)
    q3 = df3[col].quantile(0.75)
    iqr = q3 - q1
    low = q1 - 1.5*iqr
    high = q3 + 1.5*iqr
    df3 = df3[(df3[col] >= low) & (df3[col] <= high)]

# Encoding data kategorikal
# Tidak ada data kategorikal pada dataset ini

# Scaling data numerik
scaler = StandardScaler()
df3[cols] = scaler.fit_transform(df3[cols])

# Membagi dataset menjadi feature dan target
X = df3.drop('fetal_health', axis=1)
y = df3['fetal_health']
```
